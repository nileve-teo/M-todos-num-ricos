```python
import math

# ======= Convertidor de funciones =======
def convertir(expr):
    expr = expr.replace("sen", "math.sin")
    expr = expr.replace("cos", "math.cos")
    expr = expr.replace("tan", "math.tan")
    expr = expr.replace("ln", "math.log")
    expr = expr.replace("log", "math.log10")
    expr = expr.replace("e", "math.e")
    return expr

# ======= Derivada automática numérica =======
def derivada(f, x):
    h = 1e-8
    return (f(x + h) - f(x - h)) / (2*h)

# ======= Método de Newton-Raphson =======
def newton(funcion, x0, tol):
    f = lambda x: eval(funcion)

    iteraciones = 0
    print("\nIteraciones:")
    print("It\t x\t\t f(x)")

    while True:
        fx = f(x0)
        dfx = derivada(f, x0)

        print(f"{iteraciones}\t {x0:.8f}\t {fx:.2e}")

        if abs(fx) < tol:
            break

        if dfx == 0:
            print("❌ Derivada cero, no se puede continuar")
            return

        x0 = x0 - fx/dfx
        iteraciones += 1

    print("\n===== RESULTADO =====")
    print(f"Raíz aproximada: {x0:.8f}")
    print(f"f(x): {fx:.2e}")
    print(f"Iteraciones: {iteraciones}")

# ======= PROGRAMA PRINCIPAL =======
print("Método de Newton-Raphson\n")

funcion_usuario = input("Ingrese f(x): ")
funcion = convertir(funcion_usuario)

x0 = 0.5
tol = 1e-6

newton(funcion, x0, tol)
```

```python
import math

# ======= Convertidor de funciones =======
def convertir(expr):
    expr = expr.replace("sen", "math.sin")
    expr = expr.replace("cos", "math.cos")
    expr = expr.replace("tan", "math.tan")
    expr = expr.replace("ln", "math.log")
    expr = expr.replace("log", "math.log10")
    expr = expr.replace("e", "math.e")
    return expr

# ======= Método de la Secante =======
def secante(funcion, x0, x1, tol):
    f = lambda x: eval(funcion)

    iteraciones = 0
    print("\nIteraciones:")
    print("It\t x\t\t f(x)")

    while True:
        f0 = f(x0)
        f1 = f(x1)

        if (f1 - f0) == 0:
            print("❌ División por cero en la secante")
            return

        x2 = x1 - f1 * (x1 - x0) / (f1 - f0)
        f2 = f(x2)

        print(f"{iteraciones}\t {x2:.8f}\t {f2:.2e}")

        if abs(f2) < tol:
            break

        x0, x1 = x1, x2
        iteraciones += 1

    print("\n===== RESULTADO =====")
    print(f"Raíz aproximada: {x2:.8f}")
    print(f"f(x): {f2:.2e}")
    print(f"Iteraciones: {iteraciones}")

# ======= PROGRAMA PRINCIPAL =======
print("Método de la Secante\n")

funcion_usuario = input("Ingrese f(x): ")
funcion = convertir(funcion_usuario)

# Datos del problema
x0 = 0.5
x1 = 0.7
tol = 1e-6

secante(funcion, x0, x1, tol)
```

```python
import math

def f(x): return x * math.cos(x) - 1

def secante(x0, x1, tol):
    it = 0
    while True:
        it += 1
        # Fórmula de la secante
        xn = x1 - (f(x1) * (x1 - x0)) / (f(x1) - f(x0))
        error = abs(f(xn))

        print(f"Iteración {it}: xn = {xn:.8f}, |f(xn)| = {error:.2e}")

        if error < tol:
            return xn, it

        x0, x1 = x1, xn

raiz, total_it = secante(0.5, 0.7, 1e-6)
print(f"\nRESULTADO: Raíz = {raiz:.8f} en {total_it} iteraciones.")
```

# Comente:

*¿Qué método converge más rápido? ¿por qué?*

El método de Newton-Raphson es el más rápido (alrededor de 4 iteraciones), seguido por el método de la Secante.

¿Por qué?

Orden de convergencia:

*   Orden de convergencia:
El método de Newton-Raphson presenta convergencia cuadrática, lo que implica que el error disminuye muy rápidamente en cada iteración, duplicando aproximadamente el número de cifras correctas. Esto se debe a que utiliza la derivada de la función, permitiendo una mejor aproximación de la raíz en cada paso.

*   Comparación con bisección:
El método de bisección tiene convergencia lineal, ya que en cada iteración solo reduce el intervalo a la mitad. Por esta razón, requiere un mayor número de iteraciones para alcanzar la misma precisión.

*   Método de la secante:
La secante presenta convergencia superlineal (orden ≈ 1.6). Es más rápida que la bisección porque aproxima la derivada usando dos puntos, pero suele ser ligeramente más lenta que Newton-Raphson al no utilizar la derivada exacta.

```python
import sympy as sp

def resolver():
    print("--- SOLUCIONADOR DE RAÍCES (VERSIÓN ROBUSTA) ---")
    print("1. Bisección | 2. Newton-Raphson | 3. Secante | 4. Comparativo")

    try:
        opcion = int(input("\nSeleccione el número: "))
    except: return

    x = sp.symbols('x')
    nombres_permitidos = {
        'x': x, 'sin': sp.sin, 'sen': sp.sin, 'cos': sp.cos,
        'tan': sp.tan, 'exp': sp.exp, 'log': sp.log, 'ln': sp.log
    }

    print("\nIngrese f(x):")
    func_raw = input("f(x) = ").lower()

    try:
        expr = sp.sympify(func_raw, locals=nombres_permitidos)
        f_num = sp.lambdify(x, expr, 'math')
        df_expr = sp.diff(expr, x)
        df_num = sp.lambdify(x, df_expr, 'math')
    except Exception as e:
        print(f"Error: {e}")
        return

    exp_tol = int(input("Tolerancia 10^(?) (ej: -6): "))
    tol = 10**exp_tol

    if opcion in [1, 4]:
        print("\n--- BISECCIÓN ---")
        a = float(input("Límite a: "))
        b = float(input("Límite b: "))
        raiz, it = biseccion_logic(f_num, a, b, tol)
        mostrar("Bisección", raiz, it, f_num)

    if opcion in [2, 4]:
        print(f"\n--- NEWTON-RAPHSON (f' = {df_expr}) ---")
        x0 = float(input("Semilla x0: "))
        # Agregamos límites para evitar que Newton se escape al infinito
        raiz, it = newton_logic_robusto(f_num, df_num, x0, tol)
        mostrar("Newton", raiz, it, f_num)

    if opcion in [3, 4]:
        print("\n--- SECANTE ---")
        x0 = float(input("Semilla x0: "))
        x1 = float(input("Semilla x1: "))
        raiz, it = secante_logic(f_num, x0, x1, tol)
        mostrar("Secante", raiz, it, f_num)

# --- Lógica Mejorada ---

def newton_logic_robusto(f, df, x0, tol, max_it=100):
    it = 0
    xn = x0
    while it < max_it:
        try:
            derivada = df(xn)
            # Si la derivada es casi cero, Newton fallará enviando la raíz al infinito
            if abs(derivada) < 1e-10:
                print("Aviso: Derivada muy pequeña. Newton puede ser inestable.")
                return None, it

            x_siguiente = xn - f(xn) / derivada

            # Si el salto es demasiado grande (ej. más de 100 unidades), detener
            if abs(x_siguiente - xn) > 100:
                print(f"Aviso: El método está divergiendo (Salto detectado a {x_siguiente:.2f}).")
                return None, it

            xn = x_siguiente
            it += 1

            if abs(f(xn)) < tol:
                return xn, it
        except:
            return None, it

    print("Error: Se alcanzó el máximo de iteraciones.")
    return None, it

def biseccion_logic(f, a, b, tol):
    if f(a) * f(b) >= 0:
        print("Error: f(a) y f(b) deben tener signos opuestos.")
        return None, 0
    it = 0
    while (b - a) / 2 > tol:
        it += 1
        c = (a + b) / 2
        if abs(f(c)) < tol: return c, it
        if f(a) * f(c) < 0: b = c
        else: a = c
    return (a + b) / 2, it

def secante_logic(f, x0, x1, tol, max_it=100):
    it = 0
    while it < max_it:
        it += 1
        den = f(x1) - f(x0)
        if den == 0: break
        xn = x1 - (f(x1) * (x1 - x0)) / den
        if abs(f(xn)) < tol: return xn, it
        x0, x1 = x1, xn
    return x1, it

def mostrar(metodo, raiz, it, f):
    if raiz is not None:
        print(f"[{metodo}] Raíz: {raiz:.8f} | Iteraciones: {it} | Error: {abs(f(raiz)):.2e}")
    else:
        print(f"[{metodo}] No se pudo encontrar una raíz válida.")

if __name__ == "__main__":
    resolver()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def resolver():
    print("--- SOLUCIONADOR ANALÍTICO CON VISTA DE INTERVALO ---")
    print("1. Bisección | 2. Newton-Raphson | 3. Secante | 4. Comparativo")

    try:
        opcion = int(input("\nSeleccione el número: "))
    except: return

    x_sym = sp.symbols('x')
    nombres_permitidos = {'x': x_sym, 'sin': sp.sin, 'sen': sp.sin, 'cos': sp.cos, 'exp': sp.exp, 'log': sp.log, 'ln': sp.log}

    print("\nIngrese f(x) (ej: x*cos(x) - 1):")
    func_raw = input("f(x) = ").lower()

    try:
        expr = sp.sympify(func_raw, locals=nombres_permitidos)
        f_num = sp.lambdify(x_sym, expr, 'math')
        df_expr = sp.diff(expr, x_sym)
        df_num = sp.lambdify(x_sym, df_expr, 'math')
    except Exception as e:
        print(f"Error: {e}"); return

    exp_tol = int(input("Tolerancia 10^(?) (ej: -6): "))
    tol = 10**exp_tol

    # --- EJECUCIÓN Y TABLAS ---
    resultados_totales = []

    # Decidir qué pedir según la opción
    metodos = []
    if opcion in [1, 4]: metodos.append(("Bisección", "B"))
    if opcion in [2, 4]: metodos.append(("Newton-Raphson", "N"))
    if opcion in [3, 4]: metodos.append(("Secante", "S"))

    for nombre, tipo in metodos:
        print(f"\nConfiguración para {nombre}:")
        if tipo == "B":
            a, b = float(input("Límite a: ")), float(input("Límite b: "))
            r, it, hist = biseccion_logic(f_num, a, b, tol)
            limites_biseccion = (a, b) # Guardamos para la gráfica
        elif tipo == "N":
            x0 = float(input("Semilla x0: "))
            r, it, hist = newton_logic(f_num, df_num, x0, tol)
        elif tipo == "S":
            x0, x1 = float(input("Semilla x0: ")), float(input("Semilla x1: "))
            r, it, hist = secante_logic(f_num, x0, x1, tol)

        resultados_totales.append({"nombre": nombre, "raiz": r, "iters": it, "hist": hist})
        imprimir_historial(nombre, hist, f_num)

    # --- FASE DE GRÁFICAS ---
    # Gráfica 1: Panorama con Intervalo Resaltado
    if opcion in [1, 4]:
        # Si es comparativo o bisección, resaltamos el intervalo 'a' y 'b' ingresado
        graficar_panorama(f_num, func_raw, limites_biseccion)

    # Gráficas de Iteraciones
    graficar_iteraciones(f_num, resultados_totales, func_raw)

# --- Lógicas ---
def biseccion_logic(f, a, b, tol):
    if f(a) * f(b) >= 0: return None, 0, []
    it, hist = 0, []
    while (b - a) / 2 > tol:
        it += 1
        c = (a + b) / 2
        hist.append(c)
        if f(a) * f(c) < 0: b = c
        else: a = c
    return (a + b) / 2, it, hist

def newton_logic(f, df, x0, tol, max_it=50):
    it, xn, hist = 0, x0, [x0]
    while it < max_it:
        try:
            d = df(xn)
            xn_next = xn - f(xn) / d
            it += 1
            hist.append(xn_next)
            if abs(f(xn_next)) < tol: return xn_next, it, hist
            xn = xn_next
        except: break
    return None, it, hist

def secante_logic(f, x0, x1, tol, max_it=50):
    it, hist = 0, [x0, x1]
    while it < max_it:
        it += 1
        den = f(x1) - f(x0)
        if abs(den) < 1e-15: break
        xn = x1 - (f(x1) * (x1 - x0)) / den
        hist.append(xn)
        if abs(f(xn)) < tol: return xn, it, hist
        x0, x1 = x1, xn
    return x1, it, hist

def imprimir_historial(metodo, hist, f):
    print(f"\nHISTORIAL {metodo.upper()} (Primeras 5 y Últimas 5):")
    for i, val in enumerate(hist):
        if i < 5 or i > len(hist) - 6:
            print(f"It {i}: x = {val:.8f} | |f(x)| = {abs(f(val)):.2e}")

# --- Funciones de Gráfica ---
def graficar_panorama(f, titulo, limites):
    a, b = limites
    x = np.linspace(a - 5, b + 5, 1000)
    f_vec = np.vectorize(f)

    plt.figure(figsize=(10, 5))
    plt.plot(x, f_vec(x), label=f'f(x) = {titulo}', color='blue')
    plt.axhline(0, color='black', lw=1)
    plt.axvline(0, color='black', lw=1)

    # Resaltar el intervalo de búsqueda
    plt.axvspan(a, b, color='green', alpha=0.3, label=f'Intervalo [{a}, {b}]')

    plt.title("Vista General y Área de Búsqueda")
    plt.grid(True, ls='--', alpha=0.5)
    plt.legend()
    plt.show()

def graficar_iteraciones(f, resultados, titulo):
    n = len(resultados)
    fig, axes = plt.subplots(1, n, figsize=(5*n, 5), squeeze=False)
    f_vec = np.vectorize(f)

    for i, res in enumerate(resultados):
        ax = axes[0, i]
        hist = res["hist"]
        if not hist: continue

        # Zoom a la raíz
        margin = abs(max(hist) - min(hist)) * 0.5 + 0.1
        x_vals = np.linspace(min(hist) - margin, max(hist) + margin, 500)
        ax.plot(x_vals, f_vec(x_vals), color='black', alpha=0.2)
        ax.axhline(0, color='red', ls='--', lw=0.8)

        # Trayectoria
        ax.plot(hist, [f(x) for x in hist], 'o--', markersize=4, label='Pasos')
        if res["raiz"] is not None:
            ax.scatter(res["raiz"], f(res["raiz"]), color='red', s=80, zorder=5)
            ax.set_title(f"{res['nombre']}\nRaíz: {res['raiz']:.5f}")
        ax.grid(True, alpha=0.2)
        ax.legend()

    plt.suptitle(f"Análisis de Convergencia: {titulo}")
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    resolver()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def resolver():
    print("--- SOLUCIONADOR PROFESIONAL DE RAÍCES ---")
    print("1. Bisección | 2. Newton-Raphson | 3. Secante | 4. Comparativo")

    try:
        opcion = int(input("\nSeleccione el número: "))
    except: return

    x_sym = sp.symbols('x')

    nombres_permitidos = {'x': x_sym, 'sin': sp.sin, 'sen': sp.sin, 'cos': sp.cos, 'tan': sp.tan, 'exp': sp.exp, 'log': sp.log, 'ln': sp.log}

    print("\nIngrese f(x) (ej: x*cos(x) - 1):")
    func_raw = input("f(x) = ").lower()

    try:
        expr = sp.sympify(func_raw, locals=nombres_permitidos)
        f_num = sp.lambdify(x_sym, expr, 'math')
        df_expr = sp.diff(expr, x_sym)
        df_num = sp.lambdify(x_sym, df_expr, 'math')
    except Exception as e:
        print(f"Error: {e}"); return

    exp_tol = int(input("Tolerancia 10^(?) (ej: -6): "))
    tol = 10**exp_tol

    resultados_totales = []
    config_puntos = {} # Para guardar semillas y graficarlas

    # --- EJECUCIÓN ---
    ejecutar = [1] if opcion == 1 else [2] if opcion == 2 else [3] if opcion == 3 else [1, 2, 3]

    for m in ejecutar:
        if m == 1:
            print("\n[Configuración Bisección]")
            a, b = float(input("Límite a: ")), float(input("Límite b: "))
            r, it, hist = biseccion_logic(f_num, a, b, tol)
            resultados_totales.append({"nom": "Bisección", "raiz": r, "it": it, "hist": hist, "obs": "Lento pero seguro."})
            config_puntos['biseccion'] = (a, b)
        elif m == 2:
            print("\n[Configuración Newton-Raphson]")
            x0 = float(input("Semilla x0: "))
            r, it, hist = newton_logic(f_num, df_num, x0, tol)
            resultados_totales.append({"nom": "Newton-R", "raiz": r, "it": it, "hist": hist, "obs": "Convergencia cuadrática."})
            config_puntos['newton'] = x0
        elif m == 3:
            print("\n[Configuración Secante]")
            x0s, x1s = float(input("Semilla x0: ")), float(input("Semilla x1: "))
            r, it, hist = secante_logic(f_num, x0s, x1s, tol)
            resultados_totales.append({"nom": "Secante", "raiz": r, "it": it, "hist": hist, "obs": "Rápido, sin derivada."})
            config_puntos['secante'] = (x0s, x1s)

    # --- SALIDA DE DATOS ---
    if opcion in [1, 2, 3]:
        # Mostrar historial detallado para un solo método
        res = resultados_totales[0]
        imprimir_historial_detallado(res['nom'], res['hist'], f_num)
    else:
        # Mostrar cuadro comparativo resumen para la opción 4
        imprimir_cuadro_comparativo(resultados_totales, f_num)

    # --- GRÁFICAS ---
    graficar_panorama_completo(f_num, func_raw, config_puntos)
    graficar_iteraciones_separadas(f_num, resultados_totales, func_raw)

# --- Lógicas de Cálculo ---
def biseccion_logic(f, a, b, tol):
    if f(a) * f(b) >= 0: return None, 0, []
    it, hist = 0, []
    while (b - a) / 2 > tol:
        it += 1
        c = (a + b) / 2
        hist.append(c)
        if f(a) * f(c) < 0: b = c
        else: a = c
    return c, it, hist

def newton_logic(f, df, x0, tol):
    it, xn, hist = 0, x0, [x0]
    for _ in range(50):
        try:
            d = df(xn)
            xn = xn - f(xn) / d
            it += 1
            hist.append(xn)
            if abs(f(xn)) < tol: return xn, it, hist
        except: break
    return xn, it, hist

def secante_logic(f, x0, x1, tol):
    it, hist = 0, [x0, x1]
    for _ in range(50):
        it += 1
        den = f(x1) - f(x0)
        if abs(den) < 1e-15: break
        xn = x1 - (f(x1) * (x1 - x0)) / den
        hist.append(xn)
        if abs(f(xn)) < tol: return xn, it, hist
        x0, x1 = x1, xn
    return x1, it, hist

# --- Formateo de Tablas ---
def imprimir_historial_detallado(metodo, hist, f):
    print(f"\n>>> HISTORIAL DETALLADO: {metodo} <<<")
    print("-" * 60)
    print(f"{'Iter':<6} | {'Valor x':<18} | {'Error |f(x)|':<15}")
    print("-" * 60)
    for i, x in enumerate(hist):
        print(f"{i:<6} | {x:<18.10f} | {abs(f(x)):<15.2e}")
    print("-" * 60)

def imprimir_cuadro_comparativo(resultados, f):
    print("\n" + "="*85)
    print(f"{'MÉTODO':<12} | {'RAÍZ':<15} | {'IT':<4} | {'ERROR':<10} | {'OBSERVACIÓN'}")
    print("-" * 85)
    for r in resultados:
        raiz_val = f"{r['raiz']:.8f}" if r['raiz'] else "Error"
        err_val = f"{abs(f(r['raiz'])):.1e}" if r['raiz'] else "N/A"
        print(f"{r['nom']:<12} | {raiz_val:<15} | {r['it']:<4} | {err_val:<10} | {r['obs']}")
    print("="*85)

# --- Gráficas ---
def graficar_panorama_completo(f, titulo, config):
    # Rango amplio para ver la función
    x = np.linspace(-2, 5, 1000)
    f_vec = np.vectorize(f)
    plt.figure(figsize=(12, 6))
    plt.plot(x, f_vec(x), color='gray', alpha=0.5, label=f'f(x) = {titulo}')
    plt.axhline(0, color='black', lw=1)

    # Graficar puntos/áreas de cada método si existen
    if 'biseccion' in config:
        a, b = config['biseccion']
        plt.axvspan(a, b, color='green', alpha=0.2, label='Intervalo Bisección')
    if 'newton' in config:
        plt.plot(config['newton'], f(config['newton']), 'ro', label='Semilla Newton')
    if 'secante' in config:
        x0, x1 = config['secante']
        plt.plot([x0, x1], [f(x0), f(x1)], 'bx', label='Semillas Secante')

    plt.title("Vista General: Ubicación de Puntos Iniciales")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

def graficar_iteraciones_separadas(f, resultados, titulo):
    n = len(resultados)
    fig, axes = plt.subplots(1, n, figsize=(5*n, 5), squeeze=False)
    f_vec = np.vectorize(f)
    for i, res in enumerate(resultados):
        ax = axes[0, i]
        hist = res['hist']
        if not hist: continue
        margin = abs(max(hist) - min(hist)) * 0.5 + 0.1
        x_v = np.linspace(min(hist)-margin, max(hist)+margin, 500)
        ax.plot(x_v, f_vec(x_v), color='black', alpha=0.2)
        ax.axhline(0, color='red', ls='--', lw=0.8)
        ax.plot(hist, [f(x) for x in hist], 'o--', label='Pasos')
        ax.set_title(f"{res['nom']}\n({res['it']} iteraciones)")
        ax.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    resolver()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

def resolver():
    print("\n--- SOLUCIONADOR PROFESIONAL DE RAÍCES ---")
    print("1. Bisección | 2. Newton-Raphson | 3. Secante | 4. Comparativo")

    try:
        opcion = int(input("\nSeleccione el número: "))
    except: return

    x_sym = sp.symbols('x')
    nombres_permitidos = {'x': x_sym, 'sin': sp.sin, 'sen': sp.sin, 'cos': sp.cos, 'tan': sp.tan, 'exp': sp.exp, 'log': sp.log, 'ln': sp.log}

    print("\nIngrese la función f(x):")
    func_raw = input("f(x) = ").lower()

    try:
        expr = sp.sympify(func_raw, locals=nombres_permitidos)
        f_num = sp.lambdify(x_sym, expr, 'math')
        df_expr = sp.diff(expr, x_sym)
        df_num = sp.lambdify(x_sym, df_expr, 'math')
    except Exception as e:
        print(f"Error en la función: {e}"); return

    try:
        exp_tol = int(input("Tolerancia 10^(?) (ej: -6): "))
        tol = 10**exp_tol
    except: tol = 1e-6

    resultados_totales = []
    config_grafica = {}

    # --- EJECUCIÓN ---
    ejecutar = [1] if opcion == 1 else [2] if opcion == 2 else [3] if opcion == 3 else [1, 2, 3]

    for m in ejecutar:
        if m == 1:
            print("\n[Configuración Bisección]")
            a, b = float(input("Límite a: ")), float(input("Límite b: "))
            r, it, hist = biseccion_logic(f_num, a, b, tol)
            resultados_totales.append({"nom": "Bisección", "raiz": r, "it": it, "hist": hist, "obs": "Estable, requiere cambio de signo."})
            config_grafica['bisección'] = (a, b)
        elif m == 2:
            print("\n[Configuración Newton-Raphson]")
            x0 = float(input("Semilla x0: "))
            r, it, hist = newton_logic(f_num, df_num, x0, tol)
            resultados_totales.append({"nom": "Newton-R", "raiz": r, "it": it, "hist": hist, "obs": "Más rápido (Conv. Cuadrática)."})
            config_grafica['newton'] = x0
        elif m == 3:
            print("\n[Configuración Secante]")
            x0s, x1s = float(input("Semilla x0: ")), float(input("Semilla x1: "))
            r, it, hist = secante_logic(f_num, x0s, x1s, tol)
            resultados_totales.append({"nom": "Secante", "raiz": r, "it": it, "hist": hist, "obs": "Eficaz, no usa derivadas."})
            config_grafica['secante'] = (x0s, x1s)

    # --- SALIDA DE TABLAS ---
    if opcion in [1, 2, 3]:
        res = resultados_totales[0]
        imprimir_historial_detallado(res['nom'], res['hist'], f_num)
    else:
        imprimir_cuadro_comparativo(resultados_totales, f_num)

    # --- GENERACIÓN DE GRÁFICAS ---
    graficar_todo(f_num, resultados_totales, func_raw, config_grafica)

# --- Lógicas ---
def biseccion_logic(f, a, b, tol):
    if f(a) * f(b) >= 0: return None, 0, []
    it, hist = 0, []
    while (b - a) / 2 > tol:
        it += 1
        c = (a + b) / 2
        hist.append(c)
        if f(a) * f(c) < 0: b = c
        else: a = c
    return c, it, hist

def newton_logic(f, df, x0, tol):
    it, xn, hist = 0, x0, [x0]
    for _ in range(50):
        d = df(xn)
        if abs(d) < 1e-12: break
        xn = xn - f(xn) / d
        it += 1
        hist.append(xn)
        if abs(f(xn)) < tol: return xn, it, hist
    return xn, it, hist

def secante_logic(f, x0, x1, tol):
    it, hist = 0, [x0, x1]
    for _ in range(50):
        it += 1
        den = f(x1) - f(x0)
        if abs(den) < 1e-15: break
        xn = x1 - (f(x1) * (x1 - x0)) / den
        hist.append(xn)
        if abs(f(xn)) < tol: return xn, it, hist
        x0, x1 = x1, xn
    return x1, it, hist

# --- Tablas ---
def imprimir_historial_detallado(metodo, hist, f):
    print(f"\n>>> HISTORIAL COMPLETO: {metodo} <<<")
    print("-" * 65)
    print(f"{'Iter':<6} | {'Aproximación x':<20} | {'Error |f(x)|':<15}")
    print("-" * 65)
    for i, x in enumerate(hist):
        print(f"{i:<6} | {x:<20.10f} | {abs(f(x)):<15.2e}")

def imprimir_cuadro_comparativo(resultados, f):
    print("\n" + "="*95)
    print(f"{'MÉTODO':<15} | {'RAÍZ':<18} | {'IT':<4} | {'ERROR':<10} | {'OBSERVACIÓN'}")
    print("-" * 95)
    for r in resultados:
        rz = f"{r['raiz']:.8f}" if r['raiz'] is not None else "N/A"
        er = f"{abs(f(r['raiz'])):.1e}" if r['raiz'] is not None else "N/A"
        print(f"{r['nom']:<15} | {rz:<18} | {r['it']:<4} | {er:<10} | {r['obs']}")
    print("="*95)

# --- Gráficas ---
def graficar_todo(f, resultados, titulo_f, config):
    # 1. Gráfica de Panorama
    plt.figure(figsize=(10, 5))
    x_range = np.linspace(-5, 5, 1000)
    f_vec = np.vectorize(f)
    plt.plot(x_range, f_vec(x_range), color='blue', alpha=0.6, label=f'f(x) = {titulo_f}')
    plt.axhline(0, color='black', lw=1)

    if 'bisección' in config:
        plt.axvspan(config['bisección'][0], config['bisección'][1], color='green', alpha=0.2, label='Rango Bisección')
    if 'newton' in config:
        plt.plot(config['newton'], f(config['newton']), 'ro', label='Semilla Newton')
    if 'secante' in config:
        plt.plot(config['secante'], [f(config['secante'][0]), f(config['secante'][1])], 'kx', label='Semillas Secante')

    plt.title("Vista General: Función y Puntos Iniciales")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    # 2. Gráficas de Convergencia (Zoom)
    n = len(resultados)
    fig, axes = plt.subplots(1, n, figsize=(5*n, 5), squeeze=False)
    for i, res in enumerate(resultados):
        ax = axes[0, i]
        h = res['hist']
        if not h: continue
        m = abs(max(h) - min(h)) * 0.4 + 0.1
        xv = np.linspace(min(h)-m, max(h)+m, 500)
        ax.plot(xv, f_vec(xv), color='black', alpha=0.2)
        ax.axhline(0, color='red', ls='--')
        ax.plot(h, [f(x) for x in h], 'o--', markersize=4, label=f'Rutas {res["nom"]}')
        ax.set_title(f"{res['nom']}\nIT: {res['it']}")
        ax.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    resolver()
```
